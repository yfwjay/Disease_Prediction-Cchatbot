```python
# using the new dataframe we seperate the symptoms from the prognosis(disease)

# We create 2 variables
# x for symptoms
# y for the prognosis

# the symptoms dataframe
x = training_2.iloc[:,0:132]
x
# or

# features = new_training_df.iloc[:-1]
# features


# # the prognosis dataframe

y = training_2.iloc[:,-1]
y
```

```python
# since our dataframe is in continous output variables we the machine learning to use is REGRESSION

# In addition we are grouping a disease based on the symptoms entered hence we are CLASSIFYING

```

```python
# since our prognosis column is in string format we convert it to int format for consistency.

# we also need to group by prognosis so that we can reduce the number of rows but not necessary

# reduced_data = training_2.groupby("prognosis").max()
# display(reduced_data)


# we use labelling encoder a module of preprocessing to convert prognosis column to integer

# preprocessing is a module in scikitlearn

label_encoder = preprocessing.LabelEncoder()


# we fit the encoder into our column y. fit() reads the column
label_encoder.fit(y)

# we use the transform() function to convert our strings to integers

y = label_encoder.transform(y)
```

```python
# splitting our data into 70% training set and 30% testing set using the train_test_split() function


x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.3 , random_state = 42)


```

```python

# we import the regression modules
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression


# Create an Empty List and append all models, Give the Models an alias Name
# i.e we called KNeighborsRegressor()) alias KNR etc
models = []
models.append(('D Trees', DecisionTreeRegressor()))
models.append(('KNR', KNeighborsRegressor()))
models.append(('LR', LinearRegression()))
models.append(('Support Machines', SVR(gamma = 'auto')))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Gradient Boosting', GradientBoostingRegressor()))
models.append(('Support Machines', SVR(gamma = 'auto')))


# we perform coss validation on aour mos=dules to see which model is the most accurate
from sklearn.model_selection import cross_val_score, KFold

for name, model in models:
# Here we do a 10 split K-FOLD
    kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

    cv_results = cross_val_score(model , x_train , y_train , cv=kfold , scoring="r2")

    print(name, ' Results:= ', cv_results.mean())
```

```python
# from the above we can conclude that kNearest regressor is the most accurate hence we use that

model = KNeighborsRegressor()

# we fit our model into our x_train and y_train

model.fit(x_train , y_train)
```

```python
# we now predict

predictions = model.predict(x_test)
expected = y_test
print("PREDICTIONS :" , predictions)
print("EXPECTED :" ,y_test)
```

```python
from sklearn.metrics import mean_squared_error

accuracy = mean_squared_error(predictions , expected)
accuracy
```

```python
#  i read 2 csv files remaining :
# 1. symptom description.csv 
        # => when one is diagnosed with a disease the user is given a brief description of the disease
# 2. symptom precaution.csv
        # => in addition one is given safety precautions related to that disease


description_df = pd.read_csv(r"C:\Users\Admin\Desktop\MODCOM2\datascience\datsciencepdfs\project\project datasets\symptom_description.csv")
description_df

precaution_df = pd.read_csv(r"C:\Users\Admin\Desktop\MODCOM2\datascience\datsciencepdfs\project\project datasets\symptom_precaution.csv")
precaution_df
```

```python
# I create a function called predict_disease().

# The function has one parameter ie symptoms_input.The parameter is the one that prompts the user to enter symptoms they have been experiencing  and the machine predicts the disease based on users symptom


# VARIABLES INSIDE THE predict_disease() FUNCTION

    # 1. symptoms = checks if the symptoms entered by the user are in the columns of the x variable
                    # =>the x is a variable that contains symptoms/features

    # 2. input_vector = this variable converts the users input(symptoms) into a form the machine can understand ie binary
                    # =>in our case if the symptom entered by user is present it assigned a 1 else the rest of the columns are assigned 0

    # 3. predicted_label = this variable inserts the users input into our model and tries to predict the disease

    # 4. predicted_disease = the variable has the module label_encoder and the functions inverse and transform
                        # => after the machine has predicted the disease we convert the answer into human readable format ie string
                        # => we specify that the output to be rounded off to the nearest integer and converted into an integer because our label encoder expects an integer. The zero suggests we access the 1st element in our array


    # 5. description = the variable provides description of our disease diagnosed with. 
                    #=> We access the column disease in our description df and we equate it to our answer(predicted disease).This ensures we display only the rows that match our answer.
                    # => we then access the column description , convert it to numpy array and display the 1st element


    # 6. precautions = the same as description variable


# we then return 3 variables to see the disaese , it description and precautions



def predict_disease(symptoms_input):
    # Check if the input symptoms are in the training data columns
    symptoms = x.columns
    input_vector = [1 if symptom in symptoms_input else 0 for symptom in symptoms]

    # Predict the disease
    predicted_label = model.predict([input_vector])
    predicted_disease = label_encoder.inverse_transform([int(round(predicted_label[0]))])[0]

    # Get the description and precautions for the predicted disease
    description = description_df[description_df['Disease'] == predicted_disease]['Description'].values[0]
    precautions = precaution_df[precaution_df['Disease'] == predicted_disease].values[0][1:]

    return predicted_disease, description, precautions

# Example usage of the function
symptoms_input = ["scurring", 'skin_rash']
predicted_disease, description, precautions = predict_disease(symptoms_input)

print(f"Predicted Disease: {predicted_disease}")
print(f"Description: {description}")
print("Precautions:")
for precaution in precautions:
    print(f"- {precaution}")
```
